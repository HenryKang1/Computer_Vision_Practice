```python
import numpy as np
import matplotlib.pyplot as plt


x=np.arange(0,6,0.1)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label="sin")
plt.plot(x,y2,linestyle="--",label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```

```python
x=np.arange(-6,6,0.1)
y1=np.minimum(np.maximum(0,x+3),6)/6
y2=1/(1+np.exp(-x))
plt.plot(x,y1,label="H-sigmoid")
plt.plot(x,y2,linestyle="--",label="sigmoid")
plt.title('Sigmoid & H-sigmoid')
plt.legend()
plt.show()
```

```python
x=np.arange(-6,6,0.1)
a=0
y1=x*(1/(1+np.exp(-x*a))) #swish linear
b=1
y2=x*(1/(1+np.exp(-x*b))) #swish Non-linearwith negative
c=1000000
y3=x*(1/(1+np.exp(-x*c))) # swish ReLU
plt.plot(x,y1,linestyle="--",label="Beta=0")
plt.plot(x,y2,label="Beta=1")
plt.plot(x,y3,label="Beta=Inf")
plt.title('swish')
plt.legend()
plt.show()
```

```python
x=np.arange(-6,6,0.1)
y=np.maximum(0,x) # ReLU

plt.plot(x,y,"r")
plt.title('ReLU')
plt.legend()
plt.show()
```
